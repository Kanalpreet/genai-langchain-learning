```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
text_splitter.split_documents(docs)
```

text splitting technique-HTML header text splitting

```python
from langchain_text_splitters import HTMLHeaderTextSplitter


html_string = """
<!DOCTYPE html>
<html>
<head>
<title>AI Tutorial</title>
</head>
<body>

<h1>Artificial Intelligence</h1>
<p>Artificial Intelligence enables computers to perform tasks that normally require human intelligence.</p>

<h2>Machine Learning</h2>
<p>Machine Learning allows systems to learn patterns from data and make predictions.</p>

<h2>Deep Learning</h2>
<p>Deep Learning uses neural networks with many layers for image and speech recognition.</p>

<h3>Applications</h3>
<p>AI is used in healthcare, finance, education, and autonomous vehicles.</p>

<h1>Generative AI</h1>
<p>Generative AI creates text, images, music, and code.</p>

<h2>Popular Models</h2>
<p>GPT, Gemini, Claude, and Llama are popular generative AI models.</p>

</body>
</html>
"""

headers_to_split_on=[
     ("h1", "Header 1"),
     ("h2", "Header 2"),
     ("h3", "Header 3"),
     
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_split=html_splitter.split_text(html_string)
html_header_split
```

```python
import json
import requests

json_data=requests.get("https://api.github.com/repos/hwchase17/langchain/commits").json()

```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunk=json_splitter.split_json(json_data)
```
