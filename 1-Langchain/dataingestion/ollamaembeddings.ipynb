```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
)
```

```python
vectors = embeddings.embed_documents([
    "Hello, my dog is cute.",
    "Hello, my cat is cute."
])

print(len(vectors))
print(len(vectors[0]))
```
